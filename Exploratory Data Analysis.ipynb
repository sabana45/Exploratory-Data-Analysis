# Profit vs Discount

```python
sns.scatterplot(x="Profit", y = "Discount", data = df)
```

```python
sns.lineplot(y='Profit', x = 'Discount', data = df)
```

# Segment Profit & Sales

```python
df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color=['Blue', 'Orange'], figsize=(6,3))
plt.ylabel("Profit/Sales")
```

```python
plt.figure(figsize=(6,3))
sns.lineplot(x = "Quantity", y= "Profit", data=df)
```

The profit and sales are maximum in Consumer segment and minimum in Home Office segment

# Segment wise sales analysis

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Region', y = 'Sales', data = df, hue="Segment", palette="autumn")
plt.title("Segment wise sales in each region")
```

The segment wise sales in each region is somewhat similar with no important change.

# Sales in each region

```python
plt.figure(figsize=(8,4))
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

The profit and sales are maximum in West region and minimum in South region

# Profit / Loss Sales across States

```python
df1 = df.groupby('State')[['Profit', 'Sales']].sum().sort_values(by = 'Sales', ascending = False)
df1.plot.bar(color= ["Red", 'Purple'], figsize = (15,10))
plt.ylabel("Profit / Sales")
plt.title("Profit/loss and Sales across States")
```

The Sales and Profits are maximum in the States of California and New York. The States of Texas, Pennyslvania and Ohio on the other hand results in overall loss.

# Category Analysis

```python
df.groupby('Category')[['Profit', 'Sales']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(color= ["Blue", 'Orange'], figsize = (8,4))
plt.ylabel("Profit / Sales")
```

Technology and Office Supplies have high profit. Furniture have less Profit.

# Sub-Category analysis

```python
df.groupby('Sub-Category')[['Profit', 'Sales']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(color= ["Blue", 'Orange'], figsize = (8,4))
plt.ylabel("Profit / Sales")
```
